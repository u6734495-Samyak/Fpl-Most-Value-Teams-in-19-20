```
import requests

data =  requests.get('https://fantasy.premierleague.com/api/bootstrap-static/')
data =  data.json()
data.keys()
```

```
import pandas as pd
elements_df =  pd.DataFrame(data['elements'])
element_types =  pd. DataFrame(data['element_types'])
teams = pd.DataFrame(data['teams'])
elements_df.columns
```

```
usable_data =elements_df[['first_name','second_name','points_per_game','team','element_type',
'selected_by_percent','now_cost','minutes','transfers_in','value_season','total_points']]
usable_data
```

```

usable_data['position'] = usable_data.element_type.map(element_types.set_index('id').singular_name)
usable_data['Club'] = usable_data.team.map(teams.set_index('id').name)
usable_data
```

```
usable_data['value_season'] = usable_data['value_season'].astype('float32')
usable_data = usable_data.sort_values('value_season',ascending = False)
usable_data
```

```

import numpy as np
team_value = usable_data.pivot_table(index='Club',values='value_season',aggfunc=np.mean).reset_index()
team_value =  team_value.sort_values('value_season',ascending = False)
team_value
```
